```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
def mish(_input):
    return _input*torch.tanh(F.softplus(_input))
```

```python
fns = [F.relu, F.gelu, F.hardshrink, F.selu, mish]
x = torch.linspace(-4,4,200)
n = len(fns)
b = 3
fig, ax = plt.subplots(1, n ,figsize=(b*n, b))
for i, fn in enumerate(fns):
    y = fn(x)
    name = f"{fn.__name__}"
    ax[i].plot(x,y, label = name)
    ax[i].legend()
```
